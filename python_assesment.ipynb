24. Write a Python program to check whether a number is prime.


```python
num = int(input("Enter a number: "))
def check_prime(num):
    if num == 1 or num == 0:
        print("The number is not prime")
        return

    for i in range(2, int(num/2)+1):
        if num % i == 0:
            return "The number is not prime"
    return "The number is prime"

result = check_prime(num)
print(result)
```

29. Write a Python program to count word frequency in a sentence.

```python
sentence = input("Enter a sentence: ")

# Split sentence into words
words = sentence.split()

# Create empty dictionary
frequency = {}

# Count word frequency
for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Print result
print("Word Frequency:")
for word, count in frequency.items():
    print(word, ":", count)
```

47. How do you filter rows in a Pandas DataFrame?

- Data filtering is an important part of data analysis.Filtering helps extract relevant information from datasets. Pandas supports multiple filtering conditions.

1.  Boolean Indexing :- 
- This is the most standard method. You pass a boolean condition inside square brackets.

Filter for rows where Age is greater than 25
- filtered_df = df[df['Age'] > 25]

2. Using .loc[] :- 
- The .loc[] accessor is useful when you want to filter rows based on a condition and select specific columns at the same time.

Filter rows where Age > 25 and only keep 'Name' and 'City' columns
- filtered_df = df.loc[df['Age'] > 25, ['Name', 'City']]

3. Multiple Conditions:- 
- You can combine conditions using logical operators: & (AND), | (OR), and ~ (NOT). Note: Each condition must be wrapped in parentheses.

Filter for Age > 25 AND City is 'New York'
- filtered_df = df[(df['Age'] > 25) & (df['City'] == 'New York')]

4. Using .isin():- 
- Use .isin() when you want to filter rows based on a list of specific values.

Filter for rows where City is either 'New York' or 'London'
- filtered_df = df[df['City'].isin(['New York', 'London'])]

5. Using .query():- 
- The .query() method allows you to use a string-based, SQL-like syntax, which can be cleaner for complex conditions

Filter using a string expression
- filtered_df = df.query('Age > 25 and City == "New York"')

50. Write a Python program to find the factorial of a number.

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)
    
num = int(input("Enter a number to find its factorial: "))
result = factorial(num)
print("Factorial of", num, "is:", result)
```
